## File Searching

**Python `fnmatch` module provides support for Unix shell-style wildcards, which are NOT the same as regular expressions (`re` module). The special characters used in shell-style wildcards are:**

| Pattern  | Meaning                          |
| :------: | :------------------------------: |
| *        | Matches everything               |
| ?        | Matches any single character     |
| `[seq]`  | Matches any character in seq     |
| `[!seq]` | Matches any character not in seq |


**The `fnmatch.filter()` function accepts word matches or similar word matches, by using the wildcard: `(path, 'Black*')` looks for any artist with 'Black' in its name, e.g. 'Black Sabbath', 'Black Crows' etc. NOTE that it is case-insensitive.**

**NOTE: Case-sensitivity and case-preserving are different things.**

```python
import fnmatch
```

```python
def find_albums(root, artist_name):
    for path, directories, files in os.walk(root):
        
        for artist in fnmatch.filter(directories, artist_name):
            sub_folder = os.path.join(path, artist)
            
            for album_path, albums, _ in os.walk(sub_folder):
                
                for album in albums:
                    # Album name and path to songs
                    yield os.path.join(album_path, album), album
```

```python
album_list = find_albums('data/music', 'Aerosmith')

for a in album_list:
    print(a)
    #print(a[1])
    print("*" * 40)
```

```python
def find_songs(albums):
    for album in albums:
        # Every song file in album path
        for song in os.listdir(album[0]):
            yield song
```

```python
album_list = find_albums('data/music', 'Aerosmith')

song_list = find_songs(album_list)

for s in song_list:
    print(s)
```

```python
album_list = find_albums('data/music', 'Black*')

song_list = find_songs(album_list)

for s in song_list:
    print(s)
```

```python
# Same list - CASE INSENSITIVE

album_list = find_albums('data/music', 'black*')

song_list = find_songs(album_list)

for s in song_list:
    print(s)
```

## Using a Generator Expression

**Use generator expression to produce the same results. Experiment with `fnmatch` methods and assess output.**

```python
def find_albums(root, artist_name):
    caps_name = artist_name.upper()
    
    for path, directories, files in os.walk(root):
        
        # Generator expression in fnmatch method
        for artist in fnmatch.filter((d.upper() for d in directories), caps_name):
            sub_folder = os.path.join(path, artist)
            
            for album_path, albums, _ in os.walk(sub_folder):
                
                for album in albums:
                    yield os.path.join(album_path, album), album



def find_songs(albums):
    for album in albums:
        for song in os.listdir(album[0]):
            yield song
```

```python
album_list = find_albums('data/music', 'Black*')

song_list = find_songs(album_list)

for s in song_list:
    print(s)
```

```python
def find_albums(root, artist_name):
    caps_name = artist_name.upper()
    
    for path, directories, files in os.walk(root):
        
        # Generator expression with fnmatch method
        for artist in (d for d in directories if fnmatch.fnmatch(d.upper(), caps_name)):
            sub_folder = os.path.join(path, artist)
            
            for album_path, albums, _ in os.walk(sub_folder):
                
                for album in albums:
                    yield os.path.join(album_path, album), album



def find_songs(albums):
    for album in albums:
        for song in os.listdir(album[0]):
            yield song
```

```python
album_list = find_albums('data/music', 'Black*')

song_list = find_songs(album_list)

for s in song_list:
    print(s)
```

## Find files of certain extension

**In most cases, the folders and files are not so neatly organised in an operating system. You can use a generator to return the filenames for all MP3 files in a given location.**

**The generator should return all filenames of a certain extension type, which include the path to the root folder (relative path).**

```python
def dir_files_type(root, extension):
    for path, directories, files in os.walk(root):
        for file in files:
            if file.endswith(extension):
                yield os.path.join(path, file)
```

```python
dir_files_type('data/music', '.emp3')
```

```python
# Include period char when passing extension argument

for f in dir_files_type('data/music', '.emp3'):
    print(f)
```

**NOTE: To get absolute paths, use `os.path.abspath()` method.**

    absolute_path = os.path.abspath(path)
    yield os.path.join(absolute_path, file)
    
**You can also use `fnmatch.filter()` instead of the string `endswith()` method (which also reduces the lines of code):**

```python
def dir_files_type(root, extension):
    for path, directories, files in os.walk(root):
        for file in fnmatch.filter(files, '*.{}'.format(extension)):
            yield os.path.join(path, file)
```

```python
# Don't include period char when passing extension argument

for f in dir_files_type('data/music', 'emp3'):
    print(f)
```

**This is preferable to `endswith()` method, mostly because the code is simpler but you also do not need to include `.` everytime you pass an extension type, e.g. 'txt', 'pdf', 'png' etc.**
