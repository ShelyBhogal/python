# Function to play 'Fizz Buzz'

**Input a range of numbers from 1 to 100 to return the string:**

**If number is divisible by 3, return "fizz".**

**If number is divisible by 5, return "buzz".**

**If number is divisible by 3 and 5, return "fizz buzz".**

**If number is not divisible by 3 or 5, return the string number (to match the datatype of function returns).**

**ANNOTATE AND DOCUMENT THE FUNCTION!**

```python
def fizz_buzz(n: int) -> str:
    """
    Params:
        `n` is negative or positive integer 
        
    Returns: 
        String integer, or replaced with "fizz", "buzz" or 
        "fizz buzz" if number is divisible by 3, 5 or both
        respectively.
    """
    if n % 3 == 0 and n % 5 == 0:
        return "fizz buzz"
    elif n % 3 ==0:
        return "fizz"
    elif n % 5 == 0:
        return "buzz"
    else:
        return str(n)
```

```python
for i in range(1, 101):
    print(fizz_buzz(i))
```

```python
fizz_buzz(0)
```

**Create a loop to play Fizz Buzz with the computer, i.e. take it in turns to answer. The computer starts with 1 and prints 1, since it is not divisible by 3 or 5, then you get 2 and input the answer (2), then the computer gets 3 and prints "fizz", etc.**

**The game ends when there is an incorrect answer, or you reach 100.**

1. Start the computer's response (using `fizz_buzz` function)
2. Print computer response
3. Calculate answer with `fizz_buzz` function to next number
4. Input your answer
5. Compare function answer with your answer and print result
6. Repeat until loop reaches 100, where congratulations message appears

**Use a `while` loop, because the for loop causes an error**

```python
current_value = 0

while current_value < 99:
    current_value += 1
    print(fizz_buzz(current_value))
    current_value += 1
    
    answer = fizz_buzz(current_value)
    play_answer = input("Your turn? ".format(current_value))
    if play_answer != answer:
        print("WRONG! The answer is {}".format(answer))
        break
    else:
        print("CORRECT!")
        print()
        
else:
    print("Well done!! You reached {}".format(current_value))
```
